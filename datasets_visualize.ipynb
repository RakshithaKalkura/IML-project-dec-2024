# Linearly Separable Classes

In the context of machine learning and pattern recognition, linearly separable classes refer to a scenario where two or more classes of data points can be perfectly separated by a single linear boundary. This linear boundary is often referred to as a hyperplane in higher dimensions.

## Characteristics of Linearly Separable Classes

1. **Linear Boundary**: The defining characteristic of linearly separable classes is that there exists at least one straight line (in two dimensions) or hyperplane (in higher dimensions) that can separate the classes without any misclassification.
2. **Simplicity**: Linearly separable problems are simpler to solve compared to non-linearly separable problems. Algorithms like the Perceptron, Linear Support Vector Machines (SVM), and Logistic Regression can effectively find the separating hyperplane.
3. **Binary and Multiclass**: Linearly separable classes can be binary (two classes) or multiclass (more than two classes). In the case of multiclass problems, multiple hyperplanes may be required to separate all classes.

## Visualization

The following code snippet demonstrates how to plot the datasets to visualize their linear separability:

```python
plt.figure(figsize=(10, 5))
plt.plot(data1[:, 0], data1[:, 1], 'o', label='Class 1')
plt.plot(data2[:, 0], data2[:, 1], 'x', label='Class 2')
plt.plot(data3[:, 0], data3[:, 1], '^', label='Class 3')
plt.plot(data4[:, 0], data4[:, 1], 's', label='Class 4')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Datasets from Text Files')
plt.legend()
plt.show()
```

This plot will show how the different classes are distributed and how they can be separated by linear boundaries.
