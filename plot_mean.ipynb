```python
import matplotlib.pyplot as plt
import numpy as np

def load_data(file_path):
    return np.loadtxt(file_path, delimiter=' ')

# Load the training data from the file
training_data = load_data('classification_datasets/linearlySeparable/class1_train.txt')

# Calculate the mean of the training data
mean_value_x = np.mean(training_data[:, 0])
mean_value_y = np.mean(training_data[:, 1])

# Plot the training data
plt.plot(training_data[:, 0], training_data[:, 1], 'o', label='Class 1')

# Plot the mean value
plt.plot(mean_value_x, mean_value_y, '^', label='mean of Class 1')

# Add labels and legend
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Training Data of class 1 with Mean')
plt.legend()

# Show the plot
plt.show()
```
