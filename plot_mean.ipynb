
### Mean for Linearly Separable Classes

In the context of linearly separable classes, the mean of a class is a crucial statistical measure. It represents the central point of the data distribution for that class. The mean is calculated by averaging the values of each feature (dimension) across all data points in the class.

$$
\text{For a given class with data points } (x_1, y_1), (x_2, y_2), \ldots, (x_n, y_n), \text{ the mean } (\bar{x}, \bar{y}) \text{ is computed as follows:}
$$

$$
\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i
$$

$$
\bar{y} = \frac{1}{n} \sum_{i=1}^{n} y_i
$$

Where \(n\) is the number of data points in the class.




The mean provides a summary of the data's central tendency and is often used in various machine learning algorithms, such as clustering and classification. In visualizations, the mean can be plotted to show the central point of each class, helping to understand the data distribution and the separation between different classes.

In our analysis, we have calculated the mean for four different linearly separable classes and plotted them along with the training data. This helps in visualizing the central tendency of each class and understanding the overall data distribution.

```python
import matplotlib.pyplot as plt
import numpy as np

def load_data(file_path):
    return np.loadtxt(file_path, delimiter=' ')

# Load the training data from the file
training_data1 = load_data('classification_datasets/linearlySeparable/class1_train.txt')
training_data2 = load_data('classification_datasets/linearlySeparable/class2_train.txt')
training_data3 = load_data('classification_datasets/linearlySeparable/class3_train.txt')
training_data4 = load_data('classification_datasets/linearlySeparable/class4_train.txt')

# Calculate the mean of the training data of class1 
mean_value_x1 = np.mean(training_data1[:, 0])
mean_value_y1 = np.mean(training_data1[:, 1])

# Calculate the mean of the training data of class2
mean_value_x2 = np.mean(training_data2[:, 0])
mean_value_y2 = np.mean(training_data2[:, 1])

# Calculate the mean of the training data of class3
mean_value_x3 = np.mean(training_data3[:, 0])
mean_value_y3 = np.mean(training_data3[:, 1])

# Calculate the mean of the training data of class14
mean_value_x4 = np.mean(training_data4[:, 0])
mean_value_y4 = np.mean(training_data4[:, 1])

plt.figure(figsize=(10, 6)) #set plot size

# Plot the training data
plt.plot(training_data1[:, 0], training_data1[:, 1], 'o', label='Class 1', color='r')
plt.plot(training_data2[:, 0], training_data2[:, 1], '^', label='Class 2', color='b')
plt.plot(training_data3[:, 0], training_data3[:, 1], 's', label='Class 3', color='y')
plt.plot(training_data4[:, 0], training_data4[:, 1], 'x', label='Class 4', color='g')

# Plot the mean value
plt.plot(mean_value_x1, mean_value_y1, '+', label='mean of Class 1', color='black')
plt.plot(mean_value_x2, mean_value_y2, '+', label='mean of Class 2', color='black')
plt.plot(mean_value_x3, mean_value_y3, '+', label='mean of Class 3', color='black')
plt.plot(mean_value_x4, mean_value_y4, '+', label='mean of Class 4', color='black')

# Add labels and legend
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Training Data of class 1 with Mean')
plt.legend()

# Show the plot
plt.show()
```
